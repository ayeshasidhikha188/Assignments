##### 2. Write a Pandas program to find the sum, mean, max, min value of 'SepalLengthCm' column of dataframe

```python
# Sum of 'SepalLengthCm' column
sum_value = iris_df['SepalLengthCm'].sum()

# Mean of 'SepalLengthCm' column
mean_value = iris_df['SepalLengthCm'].mean()

# Maximum value of 'SepalLengthCm' column
max_value = iris_df['SepalLengthCm'].max()

# Minimum value of 'SepalLengthCm' column
min_value = iris_df['SepalLengthCm'].min()

# Display the results
print(f"Sum: {sum_value}")
print(f"Mean: {mean_value}")
print(f"Max: {max_value}")
print(f"Min: {min_value}")
```
