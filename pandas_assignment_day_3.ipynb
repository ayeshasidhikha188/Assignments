```python
import pandas as pd
import numpy as np
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")
data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

##### Write a Pandas program to create a dataframe by given dictionary and index as lables

```python
# Create a Pandas DataFrame
df = pd.DataFrame.from_dict(data, orient='columns', dtype=None)
df.index = labels  # Set the index to the given labels

# Display the DataFrame
df
```

##### 1. Write a Pandas program to sort the DataFrame first by 'name' in descending order, then by 'score' in ascending order

```python
# Create a Pandas DataFrame
df = pd.DataFrame(data, index=labels)

# Sort the DataFrame first by 'name' in descending order, then by 'score' in ascending order
sorted_df = df.sort_values(by=['name', 'score'], ascending=[False, True])

# Display the sorted DataFrame
print("DataFrame sorted by 'name' in descending order, then by 'score' in ascending order:")
sorted_df
```

###### 2. Write a Pandas program to replace the 'qualify' column contains the values 'yes' and 'no' with True and False

```python
# Replace 'yes' with True and 'no' with False in the 'qualify' column
df['qualify'] = df['qualify'].replace({'yes': True, 'no': False})

# Display the DataFrame after replacement
print("DataFrame after replacing 'qualify' column values:")
print(df)
```

##### 3. Write a Pandas program to delete the 'attempts' column from the DataFrame

```python
# Delete the 'attempts' column without specifying the column name
df = df.drop(columns=df.columns[df.columns.isin(['attempts'])])

# Display the DataFrame after deleting the column
print("DataFrame after deleting 'attempts' column:")
df
```

##### 4. Write a Pandas program to insert a new column in existing DataFrame.

```python
# Insert a new column named 'new_column' with some values
df['new_column'] = [True, False, True, False, False, True, True, False, False, True]

# Display the DataFrame after inserting the new column
print("DataFrame after inserting 'new_column':")
df
```

##### 5. Write a Pandas program to change the name 'James' to 'Maxwell' in name column of the DataFrame.

```python
# Change the name 'James' to 'Maxwell' in the 'name' column
df['name'] = df['name'].replace('James', 'Maxwell')

# Display the DataFrame after the change
print("DataFrame after changing 'James' to 'Maxwell' in 'name' column:")
df
```

##### 6. Write a Pandas program to rename columns of a given DataFrame

```python
# Rename columns
df.rename(columns={'name': 'student_name', 'score': 'test_score', 'attempts': 'num_attempts', 'qualify': 'is_qualified'}, inplace=True)

# Display the DataFrame after renaming columns
print("DataFrame after renaming columns:")
df
```

##### 7. Write a Pandas program to convert the datatype of a given column (floats to ints).

```python
# Specify the column to convert
column_to_convert = 'score'

# Check if the column exists before converting its datatype
if column_to_convert in df.columns:
    df[column_to_convert] = df[column_to_convert].astype(int, errors='ignore')

# Display the DataFrame after converting datatype
print("DataFrame after converting '{}' column datatype:".format(column_to_convert))
df
```

##### 11. Write a Pandas program to join the two given dataframes along rows and assign all data.

```python
import pandas as pd

# Given dataframes
student_data1 = pd.DataFrame({
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks': [200, 210, 190, 222, 199]
})

student_data2 = pd.DataFrame({
    'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'],
    'marks': [201, 200, 198, 219, 201]
})

# Join the dataframes along rows
all_student_data = pd.concat([student_data1, student_data2], ignore_index=True)

# Display the resulting dataframe
print("Combined Dataframe:")
all_student_data
```

##### 12. Write a Pandas program to join the two given dataframes along columns and assign all data.

```python
# Join the dataframes along columns
all_student_data = pd.merge(student_data1, student_data2, on='student_id', how='outer')

# Display the resulting dataframe
print("Combined Dataframe:")
all_student_data
```

##### 13. Write a Pandas program to join the two dataframes using the common column of both dataframes.

```python
# Join the dataframes using the common column 'student_id'
all_student_data = pd.merge(student_data1, student_data2, on='student_id')

# Display the resulting dataframe
print("Combined Dataframe:")
all_student_data
```

###### 14. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.

```python
# Left join the dataframes using keys from the left dataframe only
all_student_data_left = pd.merge(student_data1, student_data2, on='student_id', how='left')

# Display the resulting dataframe
print("Left Joined Dataframe:")
all_student_data_left
```

###### 15. Write a Pandas program to join two dataframes using keys from right dataframe only.

```python
# Right join the dataframes using keys from the right dataframe only
all_student_data_right = pd.merge(student_data1, student_data2, on='student_id', how='right')

# Display the resulting dataframe
print("Right Joined Dataframe:")
all_student_data_right
```

###### 17. Write a Pandas program to get the current date, oldest date and number of days between Current date and oldest date of Ufo dataset.

```python
import pandas as pd
from datetime import datetime
# Convert 'todays_date' and 'Date_time' columns to datetime format
ufo_data['todays_date'] = pd.to_datetime(ufo_data['todays_date'])
ufo_data['Date_time'] = pd.to_datetime(ufo_data['Date_time'], errors='coerce', format="%m/%d/%Y %H:%M")

# Get the current date
current_date = pd.to_datetime(datetime.today().date())

# Get the oldest date in the 'Date_time' column
oldest_date = ufo_data['Date_time'].min()

# Calculate the number of days between current date and oldest date
days_difference = (current_date - oldest_date).days

# Display the results
print("Current Date:", current_date)
print("Oldest Date in UFO Dataset:", oldest_date)
print("Number of Days Between Current Date and Oldest Date:", days_difference)
```

###### 18. Write a Pandas program to get all the info of the dataframe between 1950-10-10 and 1960-10-10.

```python
# Convert 'Date_time' column to datetime format
ufo_data['Date_time'] = pd.to_datetime(ufo_data['Date_time'], errors='coerce', format="%m/%d/%Y %H:%M")

# Create a mask for the date range
date_mask = (ufo_data['Date_time'] >= '1950-10-10') & (ufo_data['Date_time'] <= '1960-10-10')

# Apply the mask to filter rows within the specified date range
filtered_data = ufo_data[date_mask]

filtered_data.info()
```
