###### 5. Write a Pandas program to sort the 'cut' Series in ascending order (returns a Series) of diamonds Dataframe. 

```python
# Sort the 'cut' Series in ascending order and return a new Series
sorted_cut_series = diamonds['cut'].sort_values(ascending=True)

# Display the sorted 'cut' Series
print("Sorted 'cut' Series in ascending order:")
sorted_cut_series
```

###### 6. Write a Pandas program to sort the entire diamonds DataFrame by the 'carat' Series in ascending and descending order.

```python
# Sort the DataFrame by 'carat' in ascending order
sorted_df_asc = diamonds.sort_values(by='carat', ascending=True)

# Sort the DataFrame by 'carat' in descending order
sorted_df_desc = diamonds.sort_values(by='carat', ascending=False)

# Display the sorted DataFrames
print("DataFrame sorted by 'carat' in ascending order:")
print(sorted_df_asc)

print("\nDataFrame sorted by 'carat' in descending order:")
sorted_df_desc
```

###### 7. Write a Pandas program to filter the DataFrame rows to only show carat weight at least 0.3.

```python
# Filter the DataFrame to show rows with carat weight at least 0.3
filtered_df = diamonds[diamonds['carat'] >= 0.3]

# Display the filtered DataFrame
print("DataFrame with carat weight at least 0.3:")
filtered_df
```

###### 8.  Write a Pandas program to find the details of the diamonds where length>5, width>5 and depth>5.

```python
# Filter the DataFrame based on conditions for length, width, and depth
filtered_df = diamonds[(diamonds['x'] > 5) & (diamonds['y'] > 5) & (diamonds['z'] > 5)]

# Display the details of the filtered diamonds
print("Details of diamonds where length, width, and depth are all greater than 5:")
filtered_df
```

###### 9. Write a Pandas program to calculate the mean of each row of diamonds DataFrame.

```python
# Filter out non-numeric columns
numeric_columns = diamonds.select_dtypes(include=['float64', 'int64'])

# Calculate the mean of each row
row_means = numeric_columns.mean(axis=1)

# Add the row means as a new column to the original DataFrame
diamonds['row_mean'] = row_means

# Display the DataFrame with the new 'row_mean' column
print("DataFrame with the mean of each row:")
diamonds
```

###### 10. Write a Pandas program to calculate the mean of price for each cut  and find maximum top 3 of diamonds DataFrame. 

```python
# Calculate the mean of 'price' for each 'cut' category
cut_price_mean = diamonds.groupby('cut')['price'].mean()

# Find the top 3 maximum mean prices
top_3_max_means = cut_price_mean.nlargest(3)

# Display the mean prices for each cut and the top 3 maximum means
print("Mean prices for each cut:")
print(cut_price_mean)
print("\nTop 3 maximum mean prices:")
print(top_3_max_means)
```

##### 11. Write a Pandas program to calculate count, minimum, maximum price for each cut of diamonds DataFrame.

```python
# Calculate count, minimum, and maximum price for each 'cut' category
cut_statistics = diamonds.groupby('cut')['price'].agg(['count', 'min', 'max'])

# Display the statistics for each cut
print("Statistics for each cut:")
cut_statistics
```

###### 12. Write a Pandas program to display and count the unique values in cut series of diamonds DataFrame.

###### 13.  Write a Pandas program to count the number of missing values in each Series of diamonds DataFrame.

```python
#Count the number of missing values in each series
missing_values_count = diamonds.isnull().sum()

# Display the count of missing values for each series
print("Number of missing values in each series:")
missing_values_count
```

###### 14. Write a Pandas program to calculate the multiply of x, y and z for each cut of diamonds DataFrame.

```python
# Calculate the product of 'x', 'y', and 'z' for each 'cut' category
cut_product = diamonds.groupby('cut').apply(lambda group: group['x'] * group['y'] * group['z'])

# Display the calculated product for each cut
print("Product of 'x', 'y', and 'z' for each cut:")
print(cut_product)
```

###### 18. Write a Pandas program to get sample 75% of the diamonds DataFrame's rows without replacement and store the remaining 25% of the rows in another DataFrame.
