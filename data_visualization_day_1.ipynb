###### 1. load the iris.csv dataset.

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\mdimr\Downloads\1569615-1545342-Data_visualization\Data_visualization\Data_visualization_day_1\Iris.csv")
df.info()
```

#### 1. create a bar plot for count the frequency of the three species.

```python
# Create a bar plot for the frequency of species
species_count = df['Species'].value_counts()
species_count.plot(kind='bar', color='skyblue')

# Set plot labels and title
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.title('Frequency of Species')

# Display the plot
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
# Calculate the percentage of each species
species_percentage = df['Species'].value_counts(normalize=True) * 100

# Create a pie chart for the percentage of species
plt.pie(species_percentage, labels=species_percentage.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue', 'lightgreen'])

# Set plot title
plt.title('Percentage of Species')

# Display the plot
plt.show()
```

##### 9. Create a Heat Map

```python
# Assuming iris_df is your DataFrame
numeric_columns = iris_df.select_dtypes(include=['float64', 'int64'])

# Create a correlation matrix for numeric columns
correlation_matrix = numeric_columns.corr()

# Create a heat map
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title('Heat Map of Iris Dataset (Numeric Columns)')
plt.show()
```

###### 10. Create a distribution plot for all numerical colomns.

```python
# Assuming iris_df is your DataFrame
numeric_columns = iris_df.select_dtypes(include=['float64', 'int64'])

# Calculate the number of rows and columns for subplots
num_cols = len(numeric_columns.columns)
num_rows = (num_cols - 1) // 2 + 1

# Plot distribution for each numeric column
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns.columns, start=1):
    plt.subplot(num_rows, 2, i)
    sns.histplot(iris_df[column], kde=True)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```
