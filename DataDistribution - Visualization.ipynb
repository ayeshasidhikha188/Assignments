## 9. Stacked bar graph :
* Useful for presenting the sums of data attributes while allowing users to see how the values of these attributes contribute to the totals.

#### Examples :

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Product': ['P1', 'P2', 'P3', 'P1', 'P2', 'P3', 'P1', 'P2', 'P3'],
    'Sales': [10, 20, 30, 25, 35, 40, 45, 50, 60]
}
df = pd.DataFrame(data)
pivot_df = df.pivot_table(index='Category', columns='Product', values='Sales')
fig, ax = plt.subplots()
for product in pivot_df.columns:
    # Use 'bottom' parameter to stack the bars
    bottom = pivot_df[product].cumsum() - pivot_df[product]
    ax.barh(pivot_df.index, pivot_df[product], left=bottom, label=product)
plt.xlabel('Sales')
plt.ylabel('Categories')
plt.title('Sales Distribution by Category and Product')
plt.legend(loc='upper left')
plt.show()
```

## 2. Skewed Data :
* Skewed data refers to a dataset in which the values are not evenly distributed, resulting in a concentration of observations either towards one end of the distribution or the other.

### Types of skewed data daistribution :
* there are 3 types and they follow as : 

### i. Right - Skewed : 
* Also called positive skew, this distribution is longer on the right side of its peak than on its left.

#### Example :

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating a right skewed distribution
def right_skewed(mean, var, skew):
    u = np.random.normal(mean, var ** 0.5)
    z = np.random.normal(0, 1)
    r = skew * z + np.sqrt(1 - skew ** 2) * z
    return u * r

# Creating 1000 data points
mean = 5
var = 2
skew = 1
data = right_skewed(mean, var, skew)

# Visualizing the distribution
plt.hist(data, bins=50, alpha=0.5, color='g', label='Generated Data')
plt.title('Right Skewed Distribution')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### ii. Left - Skewed :
* Also called negative skew, this distribution is longer on the left side of its peak than on its right.

#### Example :


```python
import matplotlib.pyplot as plt
import numpy as np

# Creating a left skewed distribution
def left_skewed(mean, var, skew):
    u = np.random.normal(mean, var ** 0.5)
    z = np.random.normal(0, 1)
    r = skew * z + np.sqrt(1 - skew ** 2) * z
    return u / r

# Creating 1000 data points
mean = 5
var = 2
skew = 1
data = left_skewed(mean, var, skew)

# Visualizing the distribution
plt.hist(data, bins=50, alpha=0.5, color='g', label='Generated Data')
plt.title('Left Skewed Distribution')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### iii. Zero-Skewed :

* Less than -1 or greater than 1, the distribution is highly skewed
* Between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed

## 3. Z-Score Data Distribution :
* a statistical measurement that describes a value's relationship to the mean of a group of values.
* known as a standard score. * Z-score is measured in terms of standard deviations from the mean.
* formula for z-score is :

* Z-score is positive if the value lies above the mean.
* Z-score is negative if the value lies below the mean.
* Z-score of 0 indicates that the data point's score is identical to the mean score.



#### Example :

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(10000)
z_scores = np.zeros(data.shape)
for i in range(data.shape[0]):
    z_scores[i] = (data[i] - np.mean(data)) / np.std(data)
plt.hist(z_scores, bins=50)
plt.xlabel('Z-Scores')
plt.ylabel('Frequency')
plt.title('Z-Score Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

mean = 0
standard_deviation = 1
skewness = 0

random_samples = skewnorm.rvs(skewness, size=1000, loc=mean, scale=standard_deviation)

plt.hist(random_samples, bins=50, density=True)
plt.title('0-Skewed Distribution with Randomness')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.show()
```
